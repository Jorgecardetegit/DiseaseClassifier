```python
destination_directory = "/content/tomato_categories"
source_directory = "/content/dataset/plantvillage dataset/color"

if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)
```

```python
subdirectories = os.listdir(source_directory)
```

```python
tomato_subdirectories = [subdir for subdir in subdirectories if subdir.startswith("Tomato")]
```

```python
for subdir in tomato_subdirectories:
    source_path = os.path.join(source_directory, subdir)
    destination_path = os.path.join(destination_directory, subdir)
    shutil.copytree(source_path, destination_path)
```
